## Way to Handle Missing Value in Pandas

```
import pandas as pd
```

### Read a online file with this Command

```
sheet = pd.read_csv('https://a.uguu.se/JZ4tBQ5sUI5D_MissingValues.csv')
```

```
sheet
```

```
sheet.describe()
```

### Way to check top or bottom section of file.
Use .head() or .tail() command to check and get insight of csv file

```
sheet.head()
```

### Check How many null value is there
.isnull() is used to check how many null value is there in the file, and add .sum() to get total number of missing value in each column,

```
sheet.isnull()
```

```
sheet.isnull().sum()
```

### Discard any Missing value
Selec: how='any', if you want to delate any raw that contain any missing value, Select: how = 'all', if you want to delate raw where all value is missing.

```
NoMissingValue = sheet.dropna(how='any')
```

```
NoMissingValue.head()
```

### Performing calculation on NaN value

```
sheet*5+1
```

### Replace NaN with any value

```
sheet.fillna(-1)
```

### Fill forward
This maybe helpful for time series data.

```
sheet.fillna(method='ffill')
```
